```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "http://www.simplyhired.com/search?q=data+scientist&l=Atlanta%2C+GA"
```

```python
r=requests.get(url)
r.content
```

```python
soup = BeautifulSoup(r.content)
print soup.prettify
```

```python
data = soup.find_all('div',{"class":"card js-job"})
data
```

```python
#Scraping 1st page data from ATL
data = soup.find_all('div',{"class":"card js-job"})

df = pd.DataFrame()
for element in range(len(data)):
    title=data[element].find("h2",{"class":"serp-title"}).get_text().strip()
    company=data[element].find("span",{"class":"serp-company"}).get_text().strip()
    location=data[element].find("span",{"class":"serp-location"}).get_text().strip()
    time=data[element].find("div",{"class":"serp-timesource"}).get_text().strip()
    df=df.append({"Title":title, "Company":company, "Location": location,"Time": time},ignore_index=True)
df
```

```python
#scraping 3 pages of data from ATL
url = "http://www.simplyhired.com/search?q=data+scientist&l=atlanta%2C+ga&pn={}"

df = pd.DataFrame()
for i in range(0,4,1):
    link=url.format(i)
    r=requests.get(link)
    soup=BeautifulSoup(r.content)
    data = soup.find_all('div',{"class":"card js-job"})
    
    for element in range(len(data)):
        title=data[element].find("h2",{"class":"serp-title"}).get_text().strip()
        company=data[element].find("span",{"class":"serp-company"}).get_text().strip()
        location=data[element].find("span",{"class":"serp-location"}).get_text().strip()
        time=data[element].find("div",{"class":"serp-timesource"}).get_text().strip()
        df=df.append({"Title":title, "Company":company, "Location": location,"Time": time},ignore_index=True)
df
```

```python
#scraping data from all cities and 4 pages
url = "http://www.simplyhired.com/search?q=data+scientist&l={}&pn={}"

cities=['atlanta','charlotte','newark','chicago']
df=pd.DataFrame()
for city in cities:
    for i in range(0,3,1):
        link=url.format(city,i)
        r=requests.get(link)
        soup=BeautifulSoup(r.content)
        data = soup.find_all('div',{"class":"card js-job"})
        
        for element in range(len(data)):
            title=data[element].find("h2",{"class":"serp-title"}).get_text().strip()
            company=data[element].find("span",{"class":"serp-company"}).get_text().strip()
            location=data[element].find("span",{"class":"serp-location"}).get_text().strip()
            time=data[element].find("div",{"class":"serp-timesource"}).get_text().strip()
            df=df.append({"Title":title, "Company":company, "Location": location,"Time": time},ignore_index=True)
df
```
